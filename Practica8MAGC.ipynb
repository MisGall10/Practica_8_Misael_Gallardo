**2. Crear los siguientes gráficos con Seaborn:**

a) Mapa de Calor (Heatmap):

  * Objetivo: Visualizar la correlación entre las variables numéricas en el dataset.
  * Instrucciones: Utiliza un **mapa de calor** para mostrar la matriz de correlación entre las variables numéricas (`total_bill`, `tip, size`).
  * Asegúrate de personalizar el mapa con una paleta de colores y etiquetas en los ejes.


````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#Cargamos el data set tips en la variable df
df = sns.load_dataset('tips')
#Guardamos en un variable las columnas total_bill, tip y size
colmuns = df[['total_bill','tip','size']]
#Guardamos en otra variable la correlacion de la variable columnas previamente guardada
correlation = colmuns.corr()
#Usamo el comando heat map para hacer el mapa de calor
sns.heatmap(correlation, annot = True, cmap = 'YlGnBu', fmt = '.2f', linewidths = 1)
#Ponemos nombre a el grafico
plt.title('Matriz de correlacion entre total bill, tip y size', fontsize = 16, fontweight = 'bold')
#Ponemos titulo a el eje x
plt.xlabel("Variables Numéricas", fontsize=12, color="black")
#Ponemos titulo a el eje y
plt.ylabel("Variables Numéricas", fontsize=12, color="black")
#Imprimimos el grafico
plt.show()
````

b) Diagrama de Violín:

  * Objetivo: Mostrar la distribución de las propinas (`tip`) para diferentes días de la semana.
  * Instrucciones: Utiliza un **diagrama de violín** para comparar cómo varía la distribución de las propinas en cada uno de los días (`day`).
  * Personaliza el gráfico con colores y etiquetas adecuadas.

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#Cargamos el data set tips en la variable df
df = sns.load_dataset('tips')
#UsamoS el comando violin plot para hacer el mapa de violin
sns.violinplot(data = df, x = 'day', y = 'tip', palette = 'muted', inner="quart", scale="area")
#Ponemos nombre a el grafico
plt.title('Distibución de las propinasa por dia de la semana', fontsize = 16, fontweight = 'bold')
#Ponemos titulo a el eje x
plt.xlabel("Dia de la semana", fontsize=12, color="black")
#Ponemos titulo a el eje y
plt.ylabel("Monto de la Propina ($)", fontsize=12, color="black")
#Imprimimos el grafico
plt.show()
````

c) Gráfico de dispersión (Scatter Plot):

  * Objetivo: Analizar la relación entre el total de la cuenta (`total_bill`) y la propina (`tip`).
  * Instrucciones: Crea un gráfico de dispersión con `total_bill` en el eje X y `tip` en el eje Y. Usa un color diferente para cada día de la semana.

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#Cargamos el data set tips en la variable df
df = sns.load_dataset('tips')
#UsamoS el comando scatter plot para hacer el grafico de dispersion
sns.scatterplot(data=df, x="total_bill", y="tip", hue="day", palette="Set2", s=30, edgecolor="black")
#Ponemos nombre a el grafico
plt.title('Relacion entre toal de la cuenta y propina', fontsize = 16, fontweight = 'bold')
#Ponemos titulo a el eje x
plt.xlabel("Total de la cuenta ($)", fontsize=12, color="black")
#Ponemos titulo a el eje y
plt.ylabel("Propina ($)", fontsize=12, color="black")
#Imprimimos el grafico
plt.show()
````

d) Gráfico de barras (Bar Plot):

  * Objetivo: Comparar el promedio de la propina (`tip`) por género (`sex`).
  * Instrucciones: Utiliza un gráfico de barras para mostrar el promedio de las propinas por género (hombres y mujeres).

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#Cargamos el data set tips en la variable df
df = sns.load_dataset('tips')
#Guardamos en un variable los promedios por sexo y la propina
media = df.groupby('sex')['tip'].mean().reset_index()
#UsamoS el comando scatter plot para hacer el grafico de dispersion
sns.barplot(data=media, x="sex", y="tip", palette="Set3")
#Ponemos nombre a el grafico
plt.title('Promedio de Propinas por genero', fontsize = 16, fontweight = 'bold')
#Ponemos titulo a el eje x
plt.xlabel("Genero", fontsize=12, color="black")
#Ponemos titulo a el eje y
plt.ylabel("Promedio de propina ($)", fontsize=12, color="black")
#Imprimimos el grafico
plt.show()
````
